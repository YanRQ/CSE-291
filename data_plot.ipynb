```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [6,4]
```

Sequential

- Sequential - Pf vs. size

```python
f_sq = open("seq_result.txt", 'r')

raw_sq_list = f_sq.readlines()
seq_data = np.array([ [float(x) for x in raw_sq_list[i].strip().split(',')] for i in xrange(1, len(raw_sq_list)-1) ])
```

```python
pv1 = np.array([[item[2], item[8], item[9], item[10]] for item in seq_data if item[0] == 0 and item[2] == item[3] and item[1] == 64])
```

```python
pv2 = np.array([[item[2],item[8],item[9],item[10]] for item in seq_data if item[0]==100 and item[2] == item[3] and item[1] == 64])
```

```python
plt.plot(pv1[:,0], pv1[:,1], '-.-.', label='read')
plt.plot(pv2[:,0], pv2[:,1], 'g-', label='write')
plt.legend(loc=5)
plt.grid(True)
plt.xlabel('Read/Write Size (KB)')
plt.ylabel('Latency (usec)')
plt.show()

plt.plot(pv1[:,0], pv1[:,2], '-.-.', label='read')
plt.plot(pv2[:,0], pv2[:,2], 'g-', label='write')
plt.legend(loc=5)
plt.grid(True)
plt.xlabel('Read/Write Size (KB)')
plt.ylabel('Bandwidth (MB/s)')
plt.show()

plt.plot(pv1[:,0], pv1[:,3], '-.-.', label='read')
plt.plot(pv2[:,0], pv2[:,3], 'g-', label='write')
plt.legend(loc=5)
plt.grid(True)
plt.xlabel('Read/Write Size (KB)')
plt.ylabel('Throughput (IO/s)')
plt.show()
```

- Sequential -  Pf vs. wratio, multi size ver.

```python
pv3_1 = np.array([[item[0], item[8], item[9], item[10]] for item in seq_data if item[1]==16 and item[2] == 1 and item[3] == 1])
pv3_4 = np.array([[item[0], item[8], item[9], item[10]] for item in seq_data if item[1]==16 and item[2] == 4 and item[3] == 4])
pv3_16 = np.array([[item[0], item[8], item[9], item[10]] for item in seq_data if item[1]==16 and item[2] == 16 and item[3] == 16])
pv3_64 = np.array([[item[0], item[8], item[9], item[10]] for item in seq_data if item[1]==16 and item[2] == 64 and item[3] == 64])
pv3_256 = np.array([[item[0], item[8], item[9], item[10]] for item in seq_data if item[1]==16 and item[2] == 256 and item[3] == 256])
```

```python
plt.plot(pv3_1[:,0], pv3_1[:,1], label='size=1KB')
plt.plot(pv3_4[:,0], pv3_4[:,1], label='size=4KB')
plt.plot(pv3_16[:,0], pv3_16[:,1], label='size=16KB')
plt.plot(pv3_64[:,0], pv3_64[:,1], label='size=64KB')
plt.plot(pv3_256[:,0], pv3_256[:,1], label='size=256KB')
plt.legend(loc=7)
plt.grid(True)
plt.xlabel('Write Ratio')
plt.ylabel('Latency (usec)')
plt.show()

plt.plot(pv3_1[:,0], pv3_1[:,2], label='size=1KB')
plt.plot(pv3_4[:,0], pv3_4[:,2], label='size=4KB')
plt.plot(pv3_16[:,0], pv3_16[:,2], label='size=16KB')
plt.plot(pv3_64[:,0], pv3_64[:,2], label='size=64KB')
plt.plot(pv3_256[:,0], pv3_256[:,2], label='size=256KB')
plt.legend(loc='center left', bbox_to_anchor=(0.62, 0.35))
plt.grid(True)
plt.xlabel('Write Ratio')
plt.ylabel('Bandwidth (MB/s)')
plt.show()

plt.plot(pv3_1[:,0], pv3_1[:,3], label='size=1KB')
plt.plot(pv3_4[:,0], pv3_4[:,3], label='size=4KB')
plt.plot(pv3_16[:,0], pv3_16[:,3], label='size=16KB')
plt.plot(pv3_64[:,0], pv3_64[:,3], label='size=64KB')
plt.plot(pv3_256[:,0], pv3_256[:,3], label='size=256KB')
plt.legend(loc='upper center',bbox_to_anchor=(0.62, 1))
plt.grid(True)
plt.xlabel('Write Ratio')
plt.ylabel('Throughput (IO/s)')
plt.show()
```

Stride

- Stride - Write Pf vs. size, multi stride ver.

```python
f_std = open("stride_result.txt", 'r')

raw_std_list = f_std.readlines()
std_data = np.array([ [float(x) for x in raw_std_list[i].strip().split(',')] for i in xrange(1, len(raw_std_list)-1) ])
```

```python
pv8_4 = np.array([[item[2], item[8], item[9], item[10]] for item in std_data if item[0] == 100 and item[6] == 64 and item[7] == 0 and item[2] == item[3] and item[1] == 64 ])
pv8_16 = np.array([[item[2], item[8], item[9], item[10]] for item in std_data if item[0] == 100 and item[6] == 128 and item[7] == 0 and item[2] == item[3] and item[1] == 64 ])
pv8_64 = np.array([[item[2], item[8], item[9], item[10]] for item in std_data if item[0] == 100 and item[6] == 256 and item[7] == 0 and item[2] == item[3] and item[1] == 64 ])
```

```python
plt.plot(pv2[:,0], pv2[:,1], label='sequential')
plt.plot(pv8_4[:,0], pv8_4[:,1], label='stride 64KB')
plt.plot(pv8_16[:,0], pv8_16[:,1], label='stride 128KB')
plt.plot(pv8_64[:,0], pv8_64[:,1], label='stride 256KB')
plt.legend(loc=4)
plt.grid(True)
plt.xlabel('Read/Write Size (KB)')
plt.ylabel('Latency (usec)')
plt.show()

plt.plot(pv2[:,0], pv2[:,2], label='sequential')
plt.plot(pv8_4[:,0], pv8_4[:,2], label='stride 64KB')
plt.plot(pv8_16[:,0], pv8_16[:,2], label='stride 128KB')
plt.plot(pv8_64[:,0], pv8_64[:,2], label='stride 256KB')
plt.legend(loc=5)
plt.grid(True)
plt.xlabel('Read/Write Size (KB)')
plt.ylabel('Bandwidth (MB/s)')
plt.show()

plt.plot(pv2[:,0], pv2[:,3], label='sequential')
plt.plot(pv8_4[:,0], pv8_4[:,3], label='stride 64KB')
plt.plot(pv8_16[:,0], pv8_16[:,3], label='stride 128KB')
plt.plot(pv8_64[:,0], pv8_64[:,3], label='stride 256KB')
plt.legend(loc=5)
plt.grid(1)
plt.xlabel('Read/Write Size (KB)')
plt.ylabel('Throughput (IO/s)')
plt.show()
```
